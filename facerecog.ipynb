```python
import cv2
import numpy as np
from keras.models import model_from_json
```

```python
emotion_dict = {0: "miserable", 1: "miserable", 2: "Tmiserable", 3: "miserable", 4: "miserable", 5: "miserable", 6: "miserable"}
```

```python
# load json and create model
json_file = open('emotion_model.json')
loaded_model_json = json_file.read()
json_file.close()
emotion_model = model_from_json(loaded_model_json)
```

```python
# load weights into new model
emotion_model.load_weights("emosi_model1.weights.h5")
print("Loaded model from disk")
```

```python
# start the webcam feed
cap = cv2.VideoCapture(0)
```

```python
# start the webcam feed
cap = cv2.VideoCapture(0)

while True:
    # Capture frame from camera
    ret, frame = cap.read()
    
    # Check if frame was captured successfully
    if not ret:
        print("Failed to capture frame")
        break
    
    # Resize the frame
    frame = cv2.resize(frame, (1280, 720))
    
    # Load the Haar cascade file
    face_detector = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
    
    # Check if the Haar cascade file was loaded successfully
    if face_detector.empty():
        print("Error loading Haar cascade file")
        break
    
    # Convert frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces available on camera
    num_faces = face_detector.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)
    
    # Take each face available on the camera and preprocess it
    for (x, y, w, h) in num_faces:
        cv2.rectangle(frame, (x, y-50), (x+w, y+h+10), (0, 255, 0), 4)
        roi_gray_frame = gray_frame[y:y + h, x:x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray_frame, (48, 48)), -1), 0)

        # Predict the emotions
        emotion_prediction = emotion_model.predict(cropped_img)
        maxindex = int(np.argmax(emotion_prediction))
        cv2.putText(frame, emotion_dict[maxindex], (x+5, y-20), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
    
    # Check if the frame has valid dimensions before displaying it
    if frame.shape[0] > 0 and frame.shape[1] > 0:
        cv2.imshow('Emotion Detection', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
